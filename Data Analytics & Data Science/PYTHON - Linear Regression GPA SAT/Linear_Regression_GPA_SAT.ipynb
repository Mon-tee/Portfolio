```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('/content/Linear_Regression_GPA_SAT.csv')
```

```python
data.describe()
```

```python
y = data['GPA']
x_1 = data['SAT']
```

```python
plt.scatter(x_1,y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
x = sm.add_constant(x_1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x_1,y)
yhat =	0.0017*x_1 + 0.2750
fig = plt.plot(x_1,yhat,lw=4, c='red', label ='regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```
